```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
non = pd.read_excel("N.xlsx",index_col=False)
verb = pd.read_excel("V.xlsx", index_col=False)
adj = pd.read_excel("AV.xlsx", index_col=False)
adv = pd.read_excel("AD.xlsx", index_col=False)
conpre = pd.read_excel("CP.xlsx", index_col=False)

```

```python
#箱线图的绘制
data1 = non['hard rate'].values
data2 = verb['hard rate'].values
data3 = adj['hard rate'].values
data4 = adv['hard rate'].values
data5 = conpre['hard rate'].values
# 绘制箱线图
#plt.boxplot([data1, data2, data3, data4, data5], showfliers=False,labels=['Non','Verb','ADJ','ADV','CP'])
#ax = sns.boxplot(data=[data1, data2, data3, data4, data5],showfliers=False)
ax = sns.boxplot(data=[data1, data2, data3, data4, data5], showfliers=False)
# 设置标题和轴标签
plt.title('Boxplot of Five type word')
plt.xlabel('Five type word')
plt.ylabel('Hard Rate')
ax.set_xticklabels(['Noun', 'Verb', 'ADJ', 'ADV', 'CP'])
#plt.xticks(labels=['Non', 'Verb', 'ADJ', 'ADV', 'CP'])
plt.show()
```

```python
non_mean = np.mean(data1)
verb_mean = np.mean(data2)
adj_mean = np.mean(data3)
adv_maen = np.mean(data4)
conpre_mean = np.mean(data5)
non_std = np.std(data1)
verb_std = np.std(data2)
adj_std = np.std(data3)
adv_std = np.std(data4)
conpre_std = np.std(data5)
```

```python
#累计密度分布图绘制
#sns.kdeplot(data1,cumulative=True)
# 绘制概率直方图和概率分布曲线
sns.distplot(data1, hist=True, kde=True)
sns.distplot(data2, hist=True, kde=True)
sns.distplot(data3, hist=True, kde=True)
sns.distplot(data4, hist=True, kde=True)
sns.distplot(data5, hist=True, kde=True)
plt.title("The distrubution of every types words")
plt.ylabel("The density distribution of every types words")
plt.xlabel("The value of Hard Rate")
plt.legend(labels=['Noun', 'Verb', 'ADJ', 'ADV', 'CP'])
plt.show()
plt.savefig('1.jpg',dpi=960) # 也可以在savefig()函数中设置分辨率
```
