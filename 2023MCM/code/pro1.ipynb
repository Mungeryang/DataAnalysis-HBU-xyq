```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

```

```python
# 平滑数据预测
df = pd.read_excel("for.xlsx", parse_dates=['Date'], index_col='Date')
# fit ARIMA model
model = ARIMA(df['total'], order=(1, 2, 1))
results = model.fit()
# make predictions
forecast = results.forecast(steps=60)
forecast
forecast.to_excel("forcest23.xlsx")
```
