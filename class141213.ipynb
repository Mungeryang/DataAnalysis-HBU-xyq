分类与聚类的操作
聚类分为：
层次聚类与非层次聚类


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import dendrogram,ward
```

KMeans:
n_clusters：簇的个数，分成多少类
init：‘k-means++’;初始簇中心获取方法
n_init：获取初始簇中心迭代次数
max_iter:最大迭代次数
————————————
KMeans.fit(data)
data:DataFrame/list

```python
df = pd.read_csv("iris.csv",index_col=False)
df.dropna(inplace=True)
k_df = df.loc[:,['sepal_length','sepal_width']]
kmeans = KMeans(3).fit(k_df)
k_df['labels']=kmeans.labels_
```

```python
sns.scatterplot(data=k_df,x='sepal_length',y='sepal_width',hue='labels',style='labels',palette='Set2',)
center = pd.DataFrame(kmeans.cluster_centers_)
plt.scatter(x=center[0],y=center[1],c='red',marker='^',s=60)
plt.show()
```

```python
data = pd.read_csv("iris.csv",index_col=False)
data.dropna(inplace=True)
k_data = data.loc[:,['sepal_length','sepal_width']]
K_means = KMeans(3).fit(k_data)
k_data['lables1'] = K_means.labels_
```

```python
sns.scatterplot(data=k_data,x='sepal_length',y='sepal_width',hue='labels',style='labels1',palette='Set2',)
center1 = pd.DataFrame(K_means.cluster_centers_)
plt.scatter(x=center1[0],y=center1[1],c='red',marker='^',s=60)
plt.show()
```
