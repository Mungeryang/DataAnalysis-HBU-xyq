```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv",index_col=False,encoding="utf-8")
sns.set(style="darkgrid")
sns.jointplot(x='sepal_length',y='sepal_width',data = df,kind='kde')
plt.show()
```

```python
df = pd.read_csv("iris.csv",index_col=False,encoding="utf-8")
sns.set(style="darkgrid")
sns.jointplot(x='sepal_length',y='sepal_width',data = df,kind='scatter',hue='species')
plt.show()
```

```python
df = pd.read_csv("penguins.csv",index_col=False,encoding="utf-8")
sns.set(style="darkgrid")
sns.displot(df,x="flipper_length_mm",kind='hist',kde=True)
plt.show()
```
