Python词云可视化-jieba分词：
对于英文的分词wordcloud可以直接进行单词划分，其次单词本身有空格可以直接进行单词划分
但是对于中文不行
中文要借助jieba第三方库进行分词


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
#背景图片(必须结合numpy使用)
from PIL import Image
```

```python
with open("二十大.txt",'r',encoding='utf-8') as f:
    text = f.read()
words = jieba.lcut(text,cut_all=False)
counts = {}
excludes = {"，","、","的","。","\n",'和','—','“','”','在','是','为','了','以','要','\u3000','对','中','把','！','不','到','向','人','同','有','各','上','从','：','们','也','等','并','二','而','而','就','居','我们'}
for word in words:
    counts[word] = counts.get(word,0)+1
for word in excludes:
    del counts[word]
items = list(counts.items())
items.sort(key=lambda x:x[1],reverse=True)
for i in range(len(items)):
    word,count = items[i]
    #print("{0:<10}{1:>5}".format(word,count))
# txt = " ".join(words)
# txt
#
dic = dict(items)
bg_mask=np.array(Image.open("dang.png"))
img_color = ImageColorGenerator(bg_mask)
words = WordCloud(font_path='simhei.ttf',mask=bg_mask,background_color="white").fit_words(dic)
words.recolor(color_func=img_color)
plt.imshow(words)
plt.axis("off")
plt.show()
```
