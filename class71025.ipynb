饼图：
pie(x,explode,labels,colors,autopct,startangle)
x = [2,3,1,2]
explode:突出显示
labels：每部分的标签
colors:每部分的颜色
autopct：百分比颜色
startangle:开始角度

```python
plt.rcParams['font.sans-serif']='SimHei'
plt.figure(figsize=(8,8))
label_list = ['第一季度','第二季度','第三季度','第四季度']
size = [34,52,45,34]
color = ['b','g','r','y']
explode = [0.05,0,0,0]
plt.pie(size,colors=color,labels=label_list,explode=explode,autopct="%.2f%%")
plt.axis("equal")
plt.title("饼状图")
plt.legend(loc='upper right')
plt.show()
```

气泡图：
scatter(x,y,s,c,alpha,cmap)
s:点的大小
c:点的颜色
alpha:点的透明度
cmap：已经定义的颜色设置

```python
plt.figure(figsize=(8,8))
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30*np.random.rand(N))**2
plt.scatter(x,y,s=area,c=colors,alpha=0.7,cmap='inferno')
plt.show()
```

```python
plt.rcParams['font.sans-serif']='SimHei'
plt.figure(figsize=(8,8))
df = pd.read_csv("tips.csv",index_col=False,encoding="utf-8")
g = df.groupby("sex")["smoker"]
data = list(g)
femal = data[0][1]
mal = data[1][1]
#计数
data1 = pd.value_counts(femal)
data2 = pd.value_counts(mal)
fy = data1.iloc[0:1:,]
fn = data1.iloc[1:2:,]
my = data2.iloc[0:1:,]
mn = data2.iloc[1:2:,]
size = [fy,fn,my,mn]
labels = ['男性吸烟','男性不吸烟','女性吸烟','女性不吸烟']
color = ['b','g','r','y']
explode = [0,0,0,0]
plt.pie(size,colors=color,labels=labels,explode=explode,autopct="%.2f%%")
plt.axis("equal")
plt.title("饼状图")
plt.legend(loc='upper right')
plt.show()
```

直方图:
hist(x,bins,color,edgecolor)
x:源数据
bins：分组数量/sequence:分组依据
color:内部颜色
edgecolor:边界颜色

```python
plt.rcParams['font.sans-serif']='SimHei'
plt.figure(figsize=(8,8))
df = pd.read_csv('iris.csv',index_col=False,encoding="utf-8")
bins=np.arange(4,8,0.1)
plt.hist(df.sepal_length,bins=bins,color='r',edgecolor='b')
plt.show()
```

极坐标->雷达图:

```python
plt.rcParams['font.sans-serif']='SimHei'
plt.figure(figsize=(8,8))
#设置极坐标
plt.subplot(projection='polar')
#注意前后点的一致性
r = np.array([6,6,6,0,6,6,6])
theta = np.linspace(0.0,2*np.pi,7,endpoint=True)
labels = ['编程能力','学习能力','交往能力','性能力','异性吸引力','运动能力']
plt.plot(theta,r,color='g')
plt.fill(theta,r,color='b',alpha=0.3)
plt.thetagrids(range(0,360,60),labels)
plt.title("全能高文龙能力分析")
plt.legend()
plt.show()
```
